```python
df['Beverage'].unique()
```

```python
df['Beverage'].value_counts()
```

```python
plt.figure(figsize=(12,8),dpi=200)
sns.countplot(y=df['Beverage'], order = df['Beverage'].value_counts().index)
plt.xlabel("Count of Beverage")
plt.ylabel("Beverage")
plt.title("Beverage vs Count of Beverage")
plt.show();
```
